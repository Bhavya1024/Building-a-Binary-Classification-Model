```python
emozi_set = set()
for datapoint in train_emoticon_X:
    for emozi in datapoint:
        emozi_set.add(emozi)
emozi_list = list(emozi_set)
```

```python
# will create sparse array containing the location of each emozi
# if emozi occurs multiple times multiple arrays will be created each carrying different position of that emozi
# label is appended with the newly created sparse data

def create_location_data(X,y,emozi_list):
    arr = [[]]
    emozi_str = X
    for emozi in emozi_list:
        start = 0
        end = len(emozi_str)
        count = 0
        buffer = []
        loc_arr = []
        while((loc := emozi_str.find(emozi, start, end)) != -1):
            count+=1
            loc_arr.append(loc)
            start = loc+1
        if(count == 0):
            loc_arr = [-11]
        for array in arr:
            for loc_ in loc_arr:
                new_array = array.copy()
                new_array.append(loc_+1)
                buffer.append(new_array)
        arr = buffer.copy()
    ret_arr = [[new_X,y] for new_X in arr]
    return ret_arr
```

```python
# using position encoding

DATASET = []

for i in range(len(train_emoticon_X)):
    new_data = create_location_data(train_emoticon_X[i],train_emoticon_Y[i],emozi_list=emozi_list)
    DATASET += new_data

# random.shuffle(DATASET)
```

```python
def test_model(model, test_X, test_y): # inputs a single emozi string in test_X and its label in test_y
    prediction = [0,0]
    loc_arr = create_location_data(test_X,test_y)
    X = np.array([datapoint[0] for datapoint in loc_arr])
    y = test_y
    for encoding in X:
        res = model.predict([encoding])
        prediction[res[0]]+=1
    decision = 0 if prediction[0]>prediction[1] else 1
    if decision == y:
        return True
    else:
        return False
```

```python
def model_score(model, X_test, y_test): # inputs numpy array of all features in X_test and numpy array of all labels in y_test
    correct = 0
    incorrect = 0
    for i in range(len(X_test)):
        if(test_model(model, X_test[i], y_test[i]) == True):
            correct+=1
        else:
            incorrect+=1
    score = correct/(correct+incorrect)
    return score
```
